```python
from google.colab import drive
import glob, os
import json
import numpy as np
import matplotlib.pyplot as plt
import random

drive.mount('/content/drive')
NST_path = "/content/drive/MyDrive/NST/"
!ls $NST_path
```

### A - Gatys

```python
# gatys dir
gatys_dir = NST_path+'gatys/results-gatys-gpu'
os.chdir(gatys_dir)

times, s_loss_fin = [], []

# dict_keys(['time', 'c_loss', 's_loss'])

for file_name in glob.glob("*.json"):
  with open(file_name) as json_file:
    data = json.load(json_file)
    times.append(data["time"])
    s_loss_fin.append(data["s_loss"][-1])
t = np.array(times)
s = np.array(s_loss_fin)

print("avg time: %.2fs" % np.average(t))
print("median style loss: %.2f" % np.median(s))
```

```python
# check random example

file_name = random.choice(glob.glob("*.json"))
with open(file_name) as json_file:
    data = json.load(json_file)
#plt.plot(data["c_loss"]);

plt.plot(data["s_loss"])
plt.xlabel("Iteration")
plt.ylabel("Style loss")

plt.show()
```
