```
# code source
# https://github.com/crowsonkb/style-transfer-pytorch

import sys
import os
import random
from google.colab import drive
import IPython
import time
import json
import shutil

drive.mount('/content/drive')
NST_path = "/content/drive/MyDrive/NST/"

def listdir_nohidden(path):
    listed = []
    for f in os.listdir(path):
        if not f.startswith('.'):
            listed.append(f)
    return listed

style_names = listdir_nohidden(NST_path+"dataset/style")
content_names = listdir_nohidden(NST_path+"dataset/content")

#!nvidia-smi
!git clone https://github.com/crowsonkb/style-transfer-pytorch
!pip install -e ./style-transfer-pytorch
```

```
%%time

c_pairs = ["content14.jpg", "content16.jpg", "content22.jpg", "content25.jpg", "content27.jpg"]
s_pairs = ["witkacy2.jpg", "matejko5.jpg", "beksinski1.jpg", "wyspianski1.jpg", "witkacy5.jpg"]

for i in range(5):
  #c_name = random.choice(content_names)
  #s_name = random.choice(style_names)
  s_name = s_pairs[i]
  c_name = c_pairs[i]
  print(i+1, c_name, s_name)

  style_path = NST_path + 'dataset/style/'+s_name
  content_path = NST_path + 'dataset/content/'+c_name


  t={}; start_time = time.time()
  os.system("style_transfer "+content_path +" "+style_path+" -s 512")
  #!style_transfer $content_path $style_path -s 512
  t['total_time'] = str(time.time() - start_time)

  shutil.copyfile('out.png', NST_path+"crowson/"+c_name+"x"+s_name+".png")
  shutil.copyfile('trace.json', NST_path+"crowson/"+c_name+"x"+s_name+"-trace.json")
  with open('/content/drive/MyDrive/NST/crowson/times/'+c_name+"x"+s_name+".json", 'w') as fp:
      json.dump(t, fp)

  IPython.display.Image('out.png')
```
